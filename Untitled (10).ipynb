ADVANTAGES 

Simplicity: Flask is designed to be simple and straightforward. It has a small and easily understandable codebase, making it a great choice for beginners and experienced developers alike.

In Flask, "app routing" refers to the process of mapping URLs (Uniform Resource Locators) to specific functions in your application. This allows you to define what should happen when a user visits a particular URL.

App routing is crucial for creating dynamic web applications, as it determines which code should be executed based on the URL that the user requests. This is achieved using decorators in Flask, particularly the @app.route() decorator.

Q4





```python
#Q5

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'You can contact us at contact@example.com.'

@app.route('/user/<username>')
def user_profile(username):
    return f'Profile page of {username}'

if __name__ == '__main__':
    with app.test_request_context():
        # Using url_for to generate URLs
        print(url_for('index'))   # Output: /
        print(url_for('about'))   # Output: /about
        print(url_for('contact')) # Output: /contact
        print(url_for('user_profile', username='john_doe')) # Output: /user/john_doe
```
